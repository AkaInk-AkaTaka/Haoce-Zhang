```python
import numpy as np

# Sample data (this can be replaced with any dataset)
data = np.random.normal(loc=50, scale=10, size=100)  # generate a sample of 100 data points from a normal distribution

# Number of bootstrap samples
n_bootstrap = 1000

# Function to generate bootstrap samples and calculate the 95% confidence interval for the mean

def bootstrap_confidence_interval(data, n_bootstrap, stat_func=np.mean, confidence_level=0.95):
    """
    Generate bootstrap samples and compute confidence interval for a given statistic (e.g., mean, median).
    
    Parameters:
    data (array): The original sample data
    n_bootstrap (int): Number of bootstrap resamples
    stat_func (function): The function to compute the statistic (default: np.mean)
    confidence_level (float): The confidence level for the interval (default: 0.95)
    
    Returns:
    (lower_bound, upper_bound): The confidence interval bounds
    """
    # Generate bootstrap samples and calculate the statistic for each resample
    bootstrap_samples = np.random.choice(data, size=(n_bootstrap, len(data)), replace=True)
    bootstrap_statistics = np.apply_along_axis(stat_func, 1, bootstrap_samples)

    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.percentile(bootstrap_statistics, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_statistics, (1 + confidence_level) / 2 * 100)

    return lower_bound, upper_bound

# Calculate 95% bootstrap confidence interval for the population mean / median
mean_ci = bootstrap_confidence_interval(data, n_bootstrap, stat_func=np.mean)
median_ci = bootstrap_confidence_interval(data, n_bootstrap, stat_func=np.median)
# Output the 95% confidence interval for the mean / median
print(mean_ci)
print(median_ci)
# In Normal distribution, median is very close to mean, 
#so the outputs of 95% confidence interval for the mean and mediam do not have a large difference
```

```python
#1. What is bootstrapping?
```

```python
'''
Imagine you have a bag filled with a super large amount of  different colored balls,
but you don't know exactly how many of each color are inside.
You can take out a small handful to see the combination of colors,
but you don't have the time or resources to check every ball in the bag.
Bootstrapping is like a "guessing" method.
You repeatedly take small handfuls from the balls you've already drawn (with replacement),
and by using these "simulated" new samples, you estimate the real proportions of each color in the bag. 
This helps us learn more from a limited amount of data, similar to getting an approximation of the population characteristics by resampling.
'''
```

```python
#2. What is the main purpose of bootstrapping?
```

```python
'''
The main purpose of bootstrapping is to help us estimate the uncertainty of a statistic (like the mean) in real data,
especially when we can’t access all the data in the population. 
It allows us to simulate the process of sampling many times and estimate how much those statistics might vary in different scenarios.
This is useful for evaluating uncertainty in data.
'''
```

```python
#3. If you had a hypothesized guess about what the average of a population was, 
#   and you had a sample of size n from that population, 
#   how could you use bootstrapping to assess whether or not your guess might be plausible?
```

```python
'''
Image you think the average height of a group of people is 175 cm,
and you took a sample from that group to check your guess.
If you want to assess whether your guess is reasonable,
you can use bootstrapping. You keep “resampling” from your sample (with replacement) to simulate new datasets,
and for each resampled dataset, you calculate the mean.
After doing this many times, you’ll end up with a collection of new means.
Then, you can see how many of those simulated means are close to 175 cm.
If most of them are close to 175 cm, your guess might be plausible.
if many of them are far from 175 cm, then your guess may not be accurate.
'''
```

```python
#To quantify the effect of the vaccine, 
#I used bootstrapping,by resampling 1000 times and calculating the mean health improvement for each sample, 
#I generated a distribution of possible outcomes, which allows us to compute confidence intervals for the mean health improvement.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(100)
data = pd.read_csv('vaccine.csv')
data['HealthImprovement'] = data['FinalHealthScore'] - data['InitialHealthScore']

def bootstrap(data, num_samples=1000):
    means = []
    for _ in range(num_samples):
        sample = data.sample(frac=1, replace=True)
        means.append(sample['HealthImprovement'].mean())
    return means

boot_means = bootstrap(data)
conf_interval = np.percentile(boot_means, [2.5, 97.5])
print(f"95% Confidence Interval: {conf_interval}")
plt.figure(figsize=(10, 6))
plt.hist(boot_means, bins=20)
plt.axvline(conf_interval[0], color='red')
plt.axvline(conf_interval[1], color='red')
plt.axvline(np.mean(boot_means), color='blue')
plt.title('Bootstrapped Health Improvement Distribution')
plt.xlabel('Mean Health Improvement')
plt.ylabel('Frequency')
plt.show()
```

```python
#Findings and Discussion:
```

```python
'''
The bootstrapping analysis reveals that the mean health improvement across the sampled datasets is positive,
and the 95% confidence interval for the mean improvement does not include zero. 
Which means that the vaccine has a \positive effect on patients' health.
'''
```

```python
'''
Based on the bootstrapping analysis and the resulting confidence interval, 
I reject the null hypothesis that the vaccine has no effect on patients' health. 
The data provides evidence to suggest that the vaccine has a positive effect on health scores,
leading to improvements in patients' health after receiving the vaccine.
'''
```
