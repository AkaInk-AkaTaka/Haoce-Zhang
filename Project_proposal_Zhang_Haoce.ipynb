## II. Analysis 1: Market Factor Impact Analysis

### 2.1 Research Question
How does the market excess return (Mkt) influence individual stock returns, and is this relationship stable over time?

This research is crucial for both practitioners and academics: it enables effective risk management through beta estimation and hedging strategies, supports portfolio management decisions via market sensitivity analysis, and contributes to fundamental financial theories by testing CAPM assumptions and market efficiency.

### 2.2 Variables and Exploration Plan
- **Key Variables**:
  - Dependent: return (daily returns)
  - Independent: Mkt (market excess returns)
  
- **Visualization Plan**:
  - Scatter plot of returns vs. market returns to examine linear relationship
  - Time series plot to identify potential structural changes
  - QQ plots to check normality assumptions

*Justification*: Scatter plots will help visualize the linear relationship assumption for regression analysis, while time series plots can reveal temporal patterns.

### 2.3 Analysis Method
1. Chosen Method: Linear Regression Analysis
   Justification: 
   - Following Fama and French (1993)[1], we employ linear regression to estimate market sensitivity
   - Campbell et al. (2008)[2] show this method effectively captures time-varying relationships

### 2.4 Expected Results and Relevance
- Anticipate finding a significant positive relationship between market returns and individual returns
- Beta coefficient expected to be between 0.8 and 1.2
- Results will help understand market sensitivity of returns

## III. Analysis 2: Size and Value Effects Study

### 3.1 Research Question
Do the size (SMB) and value (HML) factors provide significant explanatory power for stock returns beyond the market factor?

### 3.2 Variables and Exploration Plan
- **Key Variables**:
  - Dependent: return
  - Independent: SMB, HML, Mkt
  
- **Visualization Plan**:
  - Correlation matrix heatmap
  - Partial regression plots
  - Time series of factor returns

*Justification*: Multiple factor analysis requires understanding of interactions between variables, which these visualizations will reveal.

### 3.3 Analysis Method
Will use multiple regression analysis:
- Model: return_t = α + β1*Mkt_t + β2*SMB_t + β3*HML_t + ε_t
- Tests:
  - Factor significance (t-tests)
  - Model fit (R-squared)
  - Multicollinearity checks (VIF)

### Expected Results and Relevance
- Expect SMB and HML to show significant explanatory power
- Anticipate improved model fit compared to market-only model
- Results will inform factor-based investment strategies
